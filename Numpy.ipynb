```python
import numpy as np
crop_yields=np.array([
    [12.5,13.2,14.1,13.8,15.2],
    [10.3,11.5,10.8,12.1,11.9],
    [15.1,14.8,16.2,15.5,16.8],
    [11.2,10.9,12.3,11.7,12.8],
])
print('crop_yields array(tons):')
print(crop_yields)
print(f'\n shape:{crop_yields.shape}')
print(f'data type:{crop_yields.dtype}')
print(f'total elements:{crop_yields.size}')
print(f'dimensions:{crop_yields.ndim}')
```

```python
avg_yield_per_farm = np.mean(crop_yields, axis=1)
print("\nAverage yield per farm:")
print(avg_yield_per_farm)
```

```python
Best_farm=np.argmax(avg_yield_per_farm)
print("\n Best performing farm:Farm",Best_farm+1)
print("their average yield:", round(avg_yield_per_farm[Best_farm], 2), "tons")
```

```python
exceptional=crop_yields>15
print(exceptional)
print('\n Exceptional yields(>15):')
print(crop_yields[exceptional])
```

```python
print('\nexceptional yields(>15):')
print(crop_yields[crop_yields > 15])
```
