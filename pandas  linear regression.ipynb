```python
import numpy as np
import pandas as pd
#create data with missing values
data=pd.DataFrame({
    'plot':['A','B','C','D','E'],
    'yield_kg':[2500,np.nan,2800,np.nan,2600],
    'rainfall':[150,200,np.nan,180,170],
    'temperature':[28,30,29,np.nan,27]

})
print("=== Data with Missing Values ===")
print(data)
#=== Detecting missing data ===
print("\n=== Detecting Missing Data ===")
print("Is null?")
print(data.isnull())
print(f"\nTotal missing per column:")
print(data.notnull().sum())
```

```python
# === Fill missing values ===
print("\n=== Filling strategies ===")
# fill with mean
data_mean=data.copy()
print(data_mean["yield_kg"])
data_mean["yield_kg"]=data_mean["yield_kg"].fillna(data_mean['yield_kg'].mean())
print('\n',"fill with mean:")
print(data_mean["yield_kg"])
```

```python
# fill with median
data_median=data.copy()
print(data_median["rainfall"])
data_median["rainfall"]=data_median["rainfall"].fillna(data_median["rainfall"].median())
print(data_median["rainfall"])
```

```python
#forward fill(use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print("\nForward fill:")
print(data_ffill)
#ffill-It replaces missing values (Nan)
# with the last valid (non -missing) value above it in the column
```

```python
#=== droping missing values ===
print("\n=== Dropping rows===")
data_droped=data.dropna()
print(data_droped)
print(f"original  rows:{len(data)}, after drop :{len(data_droped)}")
```

```python
from sklearn.metrics import classification_report

# Re-executing relevant steps to define y_test and y_pred within this cell
# This assumes Product_purchase DataFrame is available from earlier cells
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Assuming Product_purchase is loaded as in cell 'lVhgIX_Q5IdZ'
# If Product_purchase is not defined, you'll need to run cell 'lVhgIX_Q5IdZ' first
Product_purchase = pd.read_csv('Product_purchase(1).csv')

x=Product_purchase.iloc[:,[2,3]].values
y=Product_purchase.iloc[:,4].values

x_train,x_test,y_train,y_test=train_test_split(x,y ,test_size=0.25,random_state=0)

stdsc=StandardScaler()
x_train_std=stdsc.fit_transform(x_train)
x_test_std=stdsc.transform(x_test)

clf = LogisticRegression(random_state=42)
clf.fit(x_train_std, y_train)
y_pred = clf.predict(x_test_std)

print("report",classification_report(y_test,y_pred))
```

```python
# Visualising the Training set results
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
X_set, y_set = x_train_std, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
