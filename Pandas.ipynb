```python
import pandas as pd
data={
    "Farm":["Farm A","Farm B","Farm C","farm D"],
    "Crop":["Wheat","Rice","wheat","maize"],
    "area_hectare":[10,15,8,12],
    "production_tons":[30,60,24,36],
}
df=pd.DataFrame(data)
print(df)
```

```python
df.tail(2)
df.head(2)
```

```python
df.describe()
```

```python
df["yield_per_hectare"]=df["production_tons"]/df["area_hectare"]
print(df)
```

```python
df.iloc[[1,3]]
print(df)
```

```python
print(df["crops"]=="wheat")
print()
wheat_farms=df[df["crops"]=="wheat"]
print(wheat_farms)
```

```python
sorted_df=df.sort_values(by="area_hectare",ascending=False)
print(sorted_df)
```

```python
avg_yield=df.groupyby("crops")["yield_per_hectare"].sum()
print(avg_yield)
```
