```python
import matplotlib.pyplot as plt
import numpy as np
months=['jan','feb','mar','apr','may','june','july','aug','sept','oct','nov','dec']
rainfall=np.array([45,52,68,120,180,220,195,160,110,75,58,40])
temperature = np.array([18, 20, 24, 28, 32, 35, 34, 33, 30, 26, 22, 19])
wheat_yield = np.array([2.1, 2.3, 2.8, 4.2, 5.8, 6.5, 5.9, 5.2, 3.8, 2.9, 2.4, 2.0])
# Example 1: Line Plot - Rainfall Over Time
plt.figure(figsize=(10, 5))
plt.plot(months, rainfall, marker='o', linewidth=2,
         color='#1f77b4', label='Rainfall')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12)
plt.title('Monthly Rainfall Pattern', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
# plt.savefig('rainfall_pattern.png', dpi=300)
print('Figure 1: Line plot created - Monthly Rainfall Pattern')
```

```python
plt.figure(figsize=(10, 5))

# Rainfall (left Y-axis)
plt.plot(months, rainfall, marker='s', linewidth=2,
         color='blue', label='Rainfall')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12, color='blue')
plt.tick_params(axis='y', labelcolor='blue')

# Temperature (right Y-axis)
plt.twinx()     # plt.twinx() creates a second Y-axis on the right side of the same plot that shares the same X-axis.
plt.plot(months, temperature, marker='^', linewidth=2,
         color='red', label='Temperature')
plt.ylabel('Temperature (°C)', fontsize=12, color='red')
plt.tick_params(axis='y', labelcolor='red')

plt.title('Rainfall and Temperature Trends', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
# plt.tight_layout() automatically adjusts spacing between plot elements so that nothing overlaps or gets
print('Figure 2: Dual-axis plot created - Rainfall and Temperature')




```

```python
# Example 3: Bar Chart - Crop Yields by Region
regions = ['North', 'South', 'East', 'West', 'Central']
yields = [180, 228, 416, 164, 245]
colors_list = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']

plt.figure(figsize=(10, 6))
bars = plt.bar(regions, yields, color=colors_list, edgecolor='black', linewidth=1.5)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Yield (tons)', fontsize=12)
plt.title('Crop Yield by Region', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

print('Figure 3: Bar chart created - Crop Yield by Region')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(rainfall, wheat_yield, s=100, c=temperature,
            cmap='YlOrRd', edgecolors='black', linewidth=1, alpha=0.7)
plt.colorbar(label='Temperature (°C)')
plt.xlabel('Rainfall (mm)', fontsize=12)
plt.ylabel('Wheat Yield (tons/ha)', fontsize=12)
plt.title('Rainfall vs Yield Correlation (colored by temperature)',
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
print('Figure 4: Scatter plot created - Rainfall vs Yield')

```
