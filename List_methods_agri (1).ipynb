```python
# List of crops
crops = ["Rice", "Wheat", "Maize"]
# Display crops
print("Crops grown in the field:")
for i in crops:
    print(i)
```

```python

crops.append("Sugarcane") #adding new crop
print(crops)
#display the number of crops
print("Number of crops grown:", len(crops))
```

```python
if "Maize" in crops:
    print("Maize is present in the list of crops")
if "onion" in crops:
    print("onion is present in the list of crops")
else:
    print("onion is not present in the list of crops")
```

```python
crops.remove("Wheat")
print(crops)
print(len(crops))
```

```python
# Find index of a crop
index_maize = crops.index("Maize")
print("Maize is at index:", index_maize)
print(crops.index("Maize"))
```

```python
# Insert a new crop at a specific position
crops.insert(1, "Sugarcane")
print("After inserting Sugarcane:")
print(crops)
```

```python
# Count how many times Sugarcane is grown
Sugarcane_count = crops.count("Sugarcane")
print("Sugarcane is grown", Sugarcane_count, "times")
```

```python
# Remove last crop
crops.pop()
print("\nAfter removing last crop:")
print(crops)
```

```python
# Sort the crops
crops.sort()
print("Crops after sorting:")
print(crops)

# Reverse the list
crops.reverse()
print("Crops after reversing:")
print(crops)
```
